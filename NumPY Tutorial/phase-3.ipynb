```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])
```

```python
print("==== Zomato sales analysis ==== ")
print("\n Sales data shape", sales_data.shape)
print("\n Sample data for 1st 3 restau: ", sales_data[:3])
# print("\n Sample data for 1st 3 restau: ", sales_data[:, 1:])       # Here the slicing data is given in format of rows and columns.
```

```python
# total sales per year

print(np.sum(sales_data, axis=0))
yeary_total = np.sum(sales_data[:, 1:], axis=0)       # Here the value of 'axis=1' refers to calculate data set vertically (columnwise).
# yeary_total = np.sum(sales_data[:, 1:], axis=1)       # Here the value of 'axis=1' refers to calculate data set horizontally (rowwise).
print(yeary_total)
```

```python
# Minimum sales per restaurant

min_sales = np.min(sales_data[:, 1:], axis=1)
print(min_sales)
```

```python
# Maximum sales per year

max_sales = np.max(sales_data[:, 1:], axis=0)
print(max_sales)
```

```python
# Average sales per restautant

avg_sales = np.mean(sales_data[:, 1:], axis=1)
# avg_sales = np.average(sales_data[:, 1:], axis=1)     # This can also be used.
print(avg_sales)
```

```python
cumilative_sum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumilative_sum)
```

```python
plt.figure(figsize=(10,6))
plt.plot(np.mean(cumilative_sum, axis=0))
plt.title("Average cummulative sales across all restaurants")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()          # This is only used in python notebook.
```
