```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array1 = np.array([[1,2,3], [4,5,6]])
array2 = np.random.random((3,3))
array3 = np.zeros((4,4))
```

```python
np.save('array1.npy', array1)
np.save('array2.npy', array2)
np.save('array3.npy', array3)
```

```python
loaded_array1 = np.load('array1.npy')
print("Loaded array 1 : ", loaded_array1)

loaded_array2 = np.load('array2.npy')
print("Loaded array 2 : ", loaded_array2)

loaded_array3 = np.load('array3.npy')
print("Loaded array 3 : ", loaded_array3)
```

```python
try:
    loaded_logo = np.load('numpy-logo.npy')

    # Display
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(loaded_logo)
    plt.title("NumPY Logo")
    plt.grid(False)

except FileNotFoundError as e:
    print("NumPY logo is not found : ", e)
```

```python
# Changing the color of logo from light to dark.

try:
    dark_logo = 1 - loaded_logo         # Inverting the color of every pixle according to the color wheel concept.

    # Display
    plt.figure(figsize=(10, 5))
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("NumPY Dark Logo")
    plt.grid(False)

except FileNotFoundError as e:
    print("NumPY logo is not found : ", e)
```
