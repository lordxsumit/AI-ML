# Accessing Data with Pandas

```python
coffee.sample(10)       # This will give 10 random data item's.
# coffee.sample(10, random_state=1)
```

### Accessing Specific values using .loc and .iloc

```python
# Format of using .loc => coffee.loc[row, column]

# coffee.loc[0]
# coffee.loc[[0,1,2]]
# coffee.loc[[0,1,5]]
# coffee.loc[0:3]
# coffee.loc[0:]
# coffee.loc[5:]
# coffee.loc[5:8]

# coffee.loc[5:8, ["Day"]]        # Used to only fetch the 'Day' column.
coffee.loc[:, ["Day", "Units Sold"]]
```

```python
# Format of using .iloc => coffee.iloc[row, column(in index format)]

coffee.iloc[:13, [0,2]]             # It is almost same as .loc but the difference is that the upper index is excluded in .iloc and included in .loc
```

### Modification in DataFrame using .loc and .iloc

```python
coffee.iloc[1, 2] = 10      # Changed the value of first row and second column (indexwise) from 15 to 10.
coffee
```

### Accessing specific single value using .at and .iat

```python
# Format of using .at => coffee.at[row, column]

coffee.at[0, "Day"]
```

```python
# Format of using .iat => coffee.iat[row, column(in index format)]

# coffee.iat[0, 2]
# str(coffee.iat[0, 2])
coffee.iat[0, 1]
```

### Sorting of DataFrame

```python
# coffee.sort_values("Day")
# coffee.sort_values("Day", ascending=False)
coffee.sort_values(["Day", "Units Sold"], ascending=[0,1])      # This line states that sort values axxording to 'Day' and 'Units Sold' and keep ascending 'False' for 'Day' and 'True' for 'Units Sold'
```

```python
# This is another way of sorting but it has some disadvantages like -> more memory usage, low performance compared to NumPY functions.

for index, row in coffee.iterrows():
    print(index)
    print(row["Day"])
    print(row["Units Sold"])
    print("\n\n")
```

### Filtering Data

```python
# bios_data.info()
# bios_data.loc[bios_data["height_cm"] > 215]
bios_data.loc[bios_data["height_cm"] > 215, ["name", "height_cm"]]
```

```python
# Short-Hand method to write the above code.
bios_data[bios_data["height_cm"] > 215] [['name', 'height_cm']]
```

```python
# When we have multiple condtitons to apply.
bios_data[(bios_data["height_cm"] > 215) & (bios_data["born_country"] == "USA")]
```
